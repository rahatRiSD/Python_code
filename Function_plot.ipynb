```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the range for 'a' and 'c'
a_values = np.linspace(0, 5, 500)
c_values = np.linspace(0, 5, 500)

# Part (a): Plot P(X >= a) and E[X] as functions of a
prob_X_ge_a = 1 - norm.cdf(a_values)  # P(X >= a)
E_X = np.zeros_like(a_values)  # E[X] is always 0

plt.figure(figsize=(10, 6))

# Plot P(X >= a)
plt.subplot(2, 1, 1)
plt.plot(a_values, prob_X_ge_a, label=r'$P(X \geq a)$', color='blue')
plt.axhline(0, color='black',linewidth=0.5)
plt.title(r'Plot of $P(X \geq a)$ as a function of $a$')
plt.xlabel('a')
plt.ylabel('Probability')
plt.grid(True)

# Plot E[X] (constant 0)
plt.subplot(2, 1, 2)
plt.plot(a_values, E_X, label=r'$E[X]$', color='red')
plt.axhline(0, color='black',linewidth=0.5)
plt.title(r'Plot of $E[X]$ as a function of $a$')
plt.xlabel('a')
plt.ylabel('Expected Value')
plt.grid(True)

plt.tight_layout()
plt.show()

# Part (b): Plot P(|X - E[X]| >= c) and var(X)/c^2 as functions of c
prob_X_ge_c = 1 - norm.cdf(c_values)  # P(|X| >= c) = P(X >= c) + P(X <= -c)
var_X_c2 = 1 / (c_values**2)  # var(X)/c^2 = 1 / c^2 since var(X) = 1 for standard normal

plt.figure(figsize=(10, 6))

# Plot P(|X - E[X]| >= c)
plt.subplot(2, 1, 1)
plt.plot(c_values, 2 * prob_X_ge_c, label=r'$P(|X| \geq c)$', color='green')
plt.axhline(0, color='black',linewidth=0.5)
plt.title(r'Plot of $P(|X| \geq c)$ as a function of $c$')
plt.xlabel('c')
plt.ylabel('Probability')
plt.grid(True)

# Plot var(X)/c^2
plt.subplot(2, 1, 2)
plt.plot(c_values, var_X_c2, label=r'$\frac{\text{Var}(X)}{c^2}$', color='orange')
plt.axhline(0, color='black',linewidth=0.5)
plt.title(r'Plot of $\frac{\text{Var}(X)}{c^2}$ as a function of $c$')
plt.xlabel('c')
plt.ylabel('Variance divided by $c^2$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0    # Mean
sigma = 1 # Standard deviation

# Create a range of x values
x = np.linspace(-5, 5, 1000)

# Compute the normal distribution PDF using the formula
f_x = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(- (x - mu)**2 / (2 * sigma**2))

# Plot the normal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, f_x, label=r'Normal Distribution $N(\mu=0, \sigma^2=1)$', color='blue')
plt.fill_between(x, f_x, alpha=0.2, color='blue')
plt.title("Normal Distribution: Maximum Entropy", fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

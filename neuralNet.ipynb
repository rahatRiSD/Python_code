```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
# def step(x):
#     if x > 0.5:
#         return 1
#     else:
#         return 0

# # y = []
# # for i in x:
# #     y.append(step(i))
# y = list(map(step, x))

y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

```python
# def relu(x):
#     if x >= 0:
#         return x
#     else:
#         return 0
# y = list(map(relu, x))

y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

```python
alpha = 0.1

y = list(map(lambda a: a if a>=0 else a*0.1, x))
plot_graph(y,"ReLU(X)")
```
